# Importing necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import tree
# Visualize the decision tree
import matplotlib.pyplot as plt
```

# Importing the data in collab
download the dataset from : https://www.kaggle.com/datasets/mexwell/heart-disease-dataset?resource=download&select=heart_statlog_cleveland_hungary_final.csv

```python
data = pd.read_csv('/content/heart_statlog_cleveland_hungary_final.csv') #path to the dataset
# Preview the first few rows
print(data.head())
```

# Data Preprocessing

```python
# Separate the features and target variable
X = data.drop('target', axis=1)  # Features are all columns except 'target'
y = data['target']  # Target
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Model Traning

```python
# Initialize the model
clf = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42)

# Train the model
clf.fit(X_train, y_train)
```

#  Evaluate the Model

```python
# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# Data Visualization

```python
plt.figure(figsize=(20,10))
tree.plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.show()
```

# Tree Graph Explain
The decision tree graph illustrates a model used to predict whether a person has heart disease based on various characteristics, much like how a series of questions can help you narrow down a choice. Here's a simplified breakdown for a layman:

Starting Point (Root Node):
The tree begins at the top with a question about a specific characteristic, such as age or cholesterol level. This is the first decision point that helps split the data into groups.

Branches (Decisions):
Each branch from the root represents a yes or no answer to the question. For example, if the question is, "Is the age greater than 50?" one branch leads to people who are older than 50, and the other to those who are younger.

Further Questions:
As you move down the tree, there are more questions at each node, refining the groups even further based on additional characteristics, such as blood pressure or whether the individual is a smoker.

Final Outcome (Leaf Nodes):
The tree ends at leaf nodes, which provide the final prediction. Each leaf tells you the model's decision—whether the person is likely to have heart disease or not, often accompanied by how confident the model is about that decision.

Understanding Decisions:
The path from the top of the tree to a leaf node shows how the decision was made. If a person meets certain criteria (like being older and having high cholesterol), they might end up in a leaf indicating a higher likelihood of having heart disease.

In summary, this decision tree serves as a visual guide to understand how different health factors contribute to the likelihood of developing heart disease, simplifying complex medical data into a series of straightforward questions and answers.
